```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv('/content/Salary_Data (1).csv')
print(dataset)
```

```python
dataset.shape
```

```python
dataset.info
```

```python
dataset.isnull()
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.3 )
print(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
accuracy = model.score(X_test,Y_test)
print(accuracy)
```

```python
plt.scatter(X_train,Y_train,color='red' , marker='+')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary (training data)')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.show()
```

```python
WorkExperience = int(input("Experience?? "))
SalaryEarned = model.predict(np.array([[WorkExperience]]))
print('Estimated Salary ' + str(WorkExperience) + ' Years of experience are ' + str(SalaryEarned[0]))
```
